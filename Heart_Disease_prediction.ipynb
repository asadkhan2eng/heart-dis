# Elbow Method

```python
scores=[]
for i in range(1,50):
    model=KNeighborsClassifier(n_neighbors=i)
    model.fit(x_train,y_train)
    y_pred=model.predict(x_test)
    score=accuracy_score(y_test,y_pred)
    scores.append(int(score*100))
```
